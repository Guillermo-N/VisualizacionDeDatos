```python
from pandas.plotting import parallel_coordinates
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("pokemon.csv")
df.head(5)
```

## 1. Número de individuos que se incluyen dentro de cada generación de Pokemon

```python
sns.countplot(x="generation", data=df)
```

En el gráfico que se ha propuesto, el eje X representa la generación de Pokemon y el eje Y representa el número de Pokemon que se incluyen en cada generación. Al utilizar el método countplot de Seaborn, se cuenta el número de ocurrencias de cada valor en la columna "generation" del DataFrame y se muestra en un gráfico de barras.

Al examinar el gráfico, podemos inferir cuántos Pokemon hay en cada generación y comparar las diferencias entre ellas. Por ejemplo, podemos ver si algunas generaciones tienen más Pokemon que otras, o si hay una distribución equitativa de Pokemon entre las diferentes generaciones.

En cuanto a la forma de representar esta información en Matplotlib, el proceso es similar, pero en lugar de utilizar el método countplot de Seaborn, se puede utilizar el método bar de Matplotlib. La sintaxis es ligeramente diferente y se pueden personalizar algunos detalles adicionales, pero en términos generales, el resultado final sería similar.

## 2. Número de individuos (frecuencia) por cada valor de la variable “HP” en el conjunto de datos.

```python
sns.histplot(x="hp", data=df, kde= any)
```

En el histograma que se ha propuesto, el eje X representa los valores de la variable "hp" y el eje Y representa la frecuencia de ocurrencia de cada valor. Al utilizar el método histplot de Seaborn, se divide el rango de valores de "hp" en bins (intervalos) y se cuenta el número de ocurrencias de cada valor en cada bin. El resultado se muestra en un histograma, donde cada barra representa un bin y su altura indica la frecuencia de ocurrencia de los valores en ese bin.

Al examinar el histograma, podemos inferir cómo se distribuyen los valores de "hp" en el conjunto de datos y si hay algún patrón o tendencia en su distribución. Por ejemplo, podemos ver si hay más Pokemon con valores altos o bajos de "hp", o si la distribución es más uniforme.

En cuanto a otras formas de representar esta información, una opción podría ser utilizar un gráfico de densidad de kernel (también conocido como un gráfico de violín). Este gráfico muestra la distribución de los datos de forma similar al histograma, pero en lugar de mostrar barras, muestra una curva suavizada que indica la densidad de los datos en cada punto. Para obtener este gráfico en Seaborn, puedes utilizar el método violinplot y establecer el parámetro kde en True.

En cuanto a la forma de representar esta información en Matplotlib, el proceso es similar, pero en lugar de utilizar el método histplot de Seaborn, se puede utilizar el método hist de Matplotlib. La sintaxis es ligeramente diferente y se pueden personalizar algunos detalles adicionales, pero en términos generales, el resultado final sería similar.

Al examinar el gráfico, podemos inferir que la mayoría de los Pokemon tienen un peso entre 0 y 10 kg, y que hay un número menor de Pokemon legendarios que tienen un peso mayor a este rango. También se podría usar un gráfico de violín, que muestra tanto la distribución de frecuencia como la densidad de probabilidad de los datos.

Además, podemos ver que la distribución de la variable "attack" es diferente para los Pokemon legendarios y no legendarios.

# Representación Multivariable

## 1. Representar, en un solo gráfico, las relaciones entre las variables de ‘ataque’,‘defensa’ y ‘legendario’

```python
# Obtener los datos de ataque y defensa de los pokemon no legendarios
attack_non_legendary = df[df['is_legendary'] == 0]['attack']
defense_non_legendary = df[df['is_legendary'] == 0]['defense']

# Obtener los datos de ataque y defensa de los pokemon legendarios
attack_legendary = df[df['is_legendary'] == 1]['attack']
defense_legendary = df[df['is_legendary'] == 1]['defense']

# Crear el gráfico con dos llamadas a scatter para representar cada conjunto de datos
sns.scatterplot(x=attack_non_legendary, y=defense_non_legendary, marker='X')
sns.scatterplot(x=attack_legendary, y=defense_legendary, marker='o')
```

En el eje X se representa la variable de ataque, en el eje Y se representa la variable de defensa y en la 3ª dimensión se representa si un pokemon es legendario o no a través del uso de colores. Es correcta la forma de representar la 3ª dimensión, aunque otras opciones podrían ser usar etiquetas o un gráfico de barras. A partir del gráfico se puede inferir que, en general, los pokemon legendarios tienen valores de ataque y defensa más altos que los no legendarios.

## 3. Comprobar, en un solo gráfico, las relaciones entre las variables de ‘salud’, ‘ataque’, ‘ataque especial’ ‘defensa’, ‘defensa especial’ y ‘velocidad’ para el conjunto completo

```python
sns.heatmap(df[['hp', 'attack', 'sp_attack', 'defense', 'sp_defense', 'speed']].corr(), annot=True)
```

En un heatmap (o hitmap), el eje X representa las distintas variables que se están comparando y el eje Y representa los distintos valores de esas variables. En este caso, se están comparando las variables 'hp', 'attack', 'sp_attack', 'defense', 'sp_defense' y 'speed' y se están visualizando los valores de esas variables en una escala de color que va desde el azul (valores más bajos) hasta el rojo (valores más altos).

Al observar el gráfico, se pueden inferir varias conclusiones. Por ejemplo, podemos ver que la variable 'hp' tiene una relación clara con la variable 'defense' ya que se observa una zona de color rojo más oscuro en la esquina inferior derecha del gráfico, lo que indica que hay una mayor concentración de pokemon con valores altos tanto en 'hp' como en 'defense'. También se puede ver que la variable 'attack' tiene una relación clara con la variable 'sp_attack' ya que se observa una zona de color rojo más oscuro en la esquina superior derecha del gráfico, lo que indica que hay una mayor concentración de pokemon con valores altos tanto en 'attack' como en 'sp_attack'.

En general, se puede decir que existen algunas variables que tienen una relación clara entre sí, como 'hp' y 'defense' o 'attack' y 'sp_attack', mientras que otras variables no parecen tener una relación tan clara, como 'speed' y 'sp_defense'.

## 4. Comprobar, en un solo gráfico, las relaciones entre las variables de ‘ataque’, ‘ataque especial’ ‘defensa’, ‘defensa especial’, solo para aquellos pokemon de tipo 'fighting' y 'psychic'

```python
# Filtra el conjunto de datos para incluir solo pokemon de tipo 'fighting' y 'psychic'
df_filtered = df[df['type1'].isin(['fighting', 'psychic'])]

# Selecciona las columnas 'attack', 'sp_attack', 'defense', 'sp_defense' y 'type1'
df_plot = df_filtered[['attack', 'sp_attack', 'defense', 'sp_defense', 'type1']]

# Crea un diccionario de colores
colors = {'fighting': 'red', 'psychic': 'green'}

# Crea el gráfico de coordenadas paralelas
parallel_coordinates(df_plot, 'type1', color=df_filtered['type1'].map(colors))
```

En el gráfico de coordenadas paralelas se representan las variables 'attack', 'sp_attack', 'defense', 'sp_defense' en el eje X y cada individuo (pokémon) se representa en el eje Y. Cada línea en el gráfico representa un individuo y cada color representa el tipo de pokémon ('fighting' o 'psychic').

Al observar el gráfico, podemos inferir varias conclusiones. Por ejemplo, podemos ver que la mayoría de los pokémon de tipo 'fighting' tienen valores altos en las variables 'attack' y 'defense', mientras que la mayoría de los pokémon de tipo 'psychic' tienen valores más bajos en esas mismas variables. También se puede ver que hay una mayor variabilidad en los valores de las variables 'sp_attack' y 'sp_defense' para los pokémon de tipo 'psychic' en comparación con los pokémon de tipo 'fighting'.

En general, se puede decir que existen algunas variables que tienen una relación clara con el tipo de pokémon, como 'attack' y 'defense' para los pokémon de tipo 'fighting', mientras que otras variables no parecen tener una relación tan clara, como 'sp_attack' y 'sp_defense'. Además, se pueden ver características o patrones comunes para los individuos en cada grupo, como la mayoría de los pokémon de tipo 'fighting' tienen valores altos en 'attack' y 'defense' y la mayoría de los pokémon de tipo 'psychic' tienen valores más bajos en esas mismas.
