```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("pokemon.csv")
df.head(5)
```

# Representación de una variable
## 1. Número de individuos (frecuencia) de cada Tipo de Pokemon

```python
df['type1'].value_counts().plot.bar(
    figsize = (12,6),
    title = "Frecuencia de cada tipo de Pokemon",
    xlabel = "Tipo",
    ylabel = "N. Especies",
    color = ["C"+str(i) for i in range(df['type1'].value_counts().count())]
)
```

***¿Qué se representa en el eje X?***

Tipo de Pokemon.

***¿Qué se representa en el eje Y?***

Numero de Pokemon de un determinado tipo.

***¿Qué conclusiones destacadas podemos inferir a la vista del gráfico?***

Tipo de Pokemon mas abundantes y monos abundantes.

## 2. Número de individuos (frecuencia) por cada valor de la variable 'HP' en el conjunto de datos.

```python
df['hp'].value_counts().sort_index().plot.line(
    figsize = (12, 6),
    title = "Frecuencia de Pokemon por su valor HP",
    xlabel = "HP",
    ylabel = "N. Especies",
)
```

***¿Qué se representa en el eje X? ¿Qué se representa en el eje Y?***
***¿Qué conclusiones destacadas podemos inferir a la vista del gráfico?***
***¿Hay alguna otra forma de representar esto que permitiría ver los resultados más claramente?***

## 3. Frecuencia de los pokemon en función de su peso

```python
sns.histplot(
    data = df['weight_kg'], 
    bins=10
).set(
    title="Distribucion de peso de los Pokemon",
    xlabel = "Peso (Kg)",
    ylabel = "N. Especies",
)
```

# Representación de dos variables
## 1. Relación entre las capacidades de ataque y defensa

```python
df.plot.scatter(x = 'attack', y = 'defense', color = 'magenta', title = "Relaccion ataque / defensa,", figsize = (10, 6),)
```

***¿Qué se representa en el eje X?***

Los puntos de defensa.

***¿Qué se representa en el eje Y?***

Los puntos de ataque.

***¿Qué conclusiones destacadas podemos inferir a la vista del gráfico?***

Que los poquemon estan valanceados con una ligera tendencia a mayour ataque.

## 2. Relación entre las capacidades de ataque y defensa

```python
df.plot.hexbin(x = 'attack', y = 'defense', title = "Relaccion ataque / defensa", gridsize = 20, )
```

***¿Qué se representa en el eje X?***

Los puntos de defensa.

***¿Qué se representa en el eje Y?*** 

Los puntos de ataque.

***¿Qué conclusiones destacadas podemos inferir a la vista del gráfico?***

Que los pokemon estan bastante balanceados en cuanto a defensa ataque con la mayoria de los valores entorno al 50/50/

***¿Qué diferencias se aprecian si se compara con el gráfico anterior?***

El uso de un gradiente de colores para indicar el numero de ocurecias de la relaccion.

## 3. Relación entre las capacidades de ataque y defensa, agrupadas

```python
df.groupby(['is_legendary', 'generation'])[['attack', 'defense']].mean().plot.bar(stacked=True, figsize = (12, 6),)
```

***¿Qué se representa en el eje X?*** 

La cantidad de puntos de ambas habilidades.

***¿Qué se representa en el eje Y?***

Las diferentes generaciones ordenadas segmentadas por legendarios y no lengendarios.

***¿Qué conclusiones destacadas podemos inferir a la vista del gráfico?***

Que todas las generacoines tanto de lengendarios como de no legendarios estan banlanceadas encuanto a ataque y defensa.

## 4. Capacidades de cada especie en función de la generación a la que pertenece

```python
df.groupby('generation')[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']].mean().plot(figsize = (12, 6))
```

***¿Qué se representa en el eje X?***

Las distintas habilidades o estadisticas de cada pokemon.

***¿Qué se representa en el eje Y?***

Las distintas generaciones.

***¿Qué conclusiones destacadas podemos inferir a la vista del gráfico?***

Como evolucionan las distintas habilidades durante las 7 generacines.

# Subgráficas
## 1. Histograma de Ataque y de Defensa en 2 subgráficos

```python
fig, axarr = plt.pyplot.subplots(2, 1, figsize=(12, 12))

sns.histplot(
    data = df['attack'], 
    bins=10,
    ax = axarr[0]
).set(
    title="Pokemon attack rating",
    ylabel = "Frequency",
    xlabel = None
)

sns.histplot(
    data = df['defense'], 
    bins=10,
    ax = axarr[1]
).set(
    title="Pokemon defense rating",
    ylabel = "Frequency",
    xlabel = None
)
```

## 2. Gráficos mixtos en subgráfico

```python
fig, axarr = plt.pyplot.subplots(2, 2, figsize=(18, 15))
colors = ["C"+str(i) for i in range(df['type1'].value_counts().count())]

df.groupby('type1')['base_happiness'].mean().plot.bar(ax = axarr[0][0], color = colors, title = "Felicidad media por tipo")
df.groupby('type1')['percentage_male'].mean().plot.bar(ax = axarr[0][1], color = colors, title = "Porcentaje machos")
df.groupby('type1')['weight_kg'].mean().plot.bar(ax = axarr[1][0], color = colors, title = "Peso medio por tipo")
df.groupby('type1')[['base_happiness', 'percentage_male']].mean().plot.hexbin(ax = axarr[1][1], x = 'base_happiness', y = 'percentage_male', gridsize = 20)
```
